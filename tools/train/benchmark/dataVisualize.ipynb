```python
# Frequence-temperature
attributes= ["cpu0", "cpu1", "cpu2", "cpu3", "cpu4", "cpu5", "cpu6", "cpu7"]

freq_path = "./freq_data_samsung.csv"
freq_data = pd.read_csv(freq_path, index_col=0)

x = freq_data["stamp"]
y1 = freq_data["temperature"]

x_sample = [i * 100 for i in x] [:len(x)//100]

y1 = y1[0:-(len(y1)%100)] #只用整百倍数个数据点
y1_sample = [0]*(len(y1)//100)

for i in range(len(y1)//100):
    y1_sample[i] = sum(y1[i*100:(i+1)*100])/100
#print(y1)

for cpu in attributes[:]:
    y2 = freq_data[cpu]
    
    y2 = y2[0:-(len(y2)%100)] #只用整百倍数个数据点
    #print(y2)
    
    y2_sample = [0]*(len(y2)//100)

    for i in range(len(y2)//100):
        y2_sample[i] = sum(y2[i*100:(i+1)*100])/100

    fig = plt.figure(figsize=(10,5))
    plt.title("cpu")

    ax1 = fig.add_subplot(111)
    ax1.plot(x_sample, y1_sample, color="r", label="temperature")
    ax1.set_ylabel('temperature')
    ax1.set_xlabel('time/ms')
    ax1.set_title(cpu)
    plt.legend()
    
    ax2 = ax1.twinx()  # this is the important function
    ax2.plot(x_sample, y2_sample, 'r', color="k", label="Freq")
    ax2.set_ylabel('Freq')
    ax2.set_xlabel('time/ms')

    plt.legend()
    #plt.hlines(1500, 0.001, 60, linestyles='dashed',linewidth=0.9,colors='#3C323C') # 画水平虚线
    

    


```
